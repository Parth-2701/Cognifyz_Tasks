To determine if there is any correlation between the restaurants location and its rating 

```python
correlation = df['Aggregate rating'].corr(df['Country Code'])
correlation 
#Creating a scatter plot to visualise this 
plt.figure(figsize = (10,6))
sns.scatterplot(x='Country Code',y='Aggregate rating',data = df,alpha =0.5)
plt.title('Relation between Aggregate rating amd Location')
plt.xlabel('Country Code')
plt.ylabel('Rating')
plt.show()
```
